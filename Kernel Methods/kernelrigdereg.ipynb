```python
# Kernel Ridge Regression with a single alpha (regularization)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.kernel_ridge import KernelRidge

# -----------------------------
# 1. Training data
# -----------------------------
X = np.array([1, 2, 3]).reshape(-1, 1)  # shape (n_samples, 1)
y = np.array([2, 3, 2.5])               # target values

# Points where we want predictions
x_star = np.linspace(0, 4, 100).reshape(-1, 1)

# -----------------------------
# 2. Parameters
# -----------------------------
l = 1.0                # length scale of RBF
gamma = 1 / (2 * l**2) # convert length scale to sklearn gamma
alpha_val = 0.1        # single regularization value

# -----------------------------
# 3. Create and fit the Kernel Ridge model
# -----------------------------
model = KernelRidge(alpha=alpha_val, kernel='rbf', gamma=gamma)
model.fit(X, y)

# -----------------------------
# 4. Predict at new points
# -----------------------------
y_pred = model.predict(x_star)

# -----------------------------
# 5. Plot training points and prediction
# -----------------------------
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='red', label='Training points', s=50)
plt.plot(x_star, y_pred, label=f'Prediction (alpha={alpha_val})', color='blue')
plt.title("Kernel Ridge Regression with RBF Kernel")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# -----------------------------
# 6. Predict a single new point x* = 1.5
# -----------------------------
x_new = np.array([1.5]).reshape(-1, 1)
y_new_pred = model.predict(x_new)
print(f"Prediction at x* = {x_new[0,0]} is {y_new_pred[0]:.4f}")

# -----------------------------
# 7. Access internal dual coefficients
# -----------------------------
dual_coefs = model.dual_coef_
print("Dual coefficients (alpha vector) =", dual_coefs)
```
